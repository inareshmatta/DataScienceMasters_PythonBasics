### Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a web framework for Python that allows you to build web applications quickly and easily. It is lightweight and flexible, which makes it a popular choice for developers who want to create simple to complex web applications.

Here are some of the advantages of using Flask:

Lightweight: Flask is a lightweight framework that doesn't have any specific tools or libraries. This means that you can build your application with only the components that you need.

Flexibility: Flask is very flexible, which means that you can easily customize it to suit your needs. It allows you to choose the components that you want to use, and you can easily add or remove features as needed.

Simple to Use: Flask is very easy to use, even for developers who are new to web development. It has a simple API and doesn't require a lot of setup or configuration.

Easy to Learn: Flask is a beginner-friendly framework that is easy to learn. It has a clear and concise documentation, and there are many tutorials and resources available online.

Large Community: Flask has a large and active community of developers who are constantly improving and updating the framework. This means that you can easily find help and support when you need it.

Compatibility: Flask is compatible with many other tools and libraries, which makes it easy to integrate with your existing infrastructure.

Overall, Flask is a great choice for developers who want to build web applications quickly and easily, without sacrificing flexibility and control over the development process.

### Q3. What is App routing in Flask? Why do we use app routes?

In Flask, routing refers to the process of directing incoming HTTP requests to the appropriate view function, which generates an HTTP response that is sent back to the client. App routing in Flask refers to defining the URL paths for various pages or endpoints of a Flask application.

Flask uses the @app.route() decorator to define app routes. The decorator specifies the URL path for a particular view function that should be executed when that URL is requested. For example, consider the following code:

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to my Flask App!'

@app.route('/about')
def about():
    return 'This is a Flask App about page.'

if __name__ == '__main__':
    app.run(debug=True)
```

Here, we have defined two app routes: '/' and '/about'. When a user visits the root URL ('/'), Flask calls the home() function and returns the message "Welcome to my Flask App!". Similarly, when a user visits the URL '/about', Flask calls the about() function and returns the message "This is a Flask App about page.".

App routing is essential in Flask because it allows developers to create a well-organized and scalable web application. By defining app routes, Flask can match incoming requests to the appropriate view function, making it easy to manage different parts of the application. Additionally, Flask's app routing system makes it easy to modify or add new routes to an existing application.

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/hello/<name>')
def hello(name):
    return f'Hello, {name}!'

with app.test_request_context():
    # Use url_for() to generate a URL to the 'hello' function
    url = url_for('hello', name='John')
    print(url)
```
