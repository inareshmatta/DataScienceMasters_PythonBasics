### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

**Difference Between a t-test and a z-test:**

1. **Standard Deviation Known vs. Unknown:**
   - **Z-Test:** Assumes that the population standard deviation $(\sigma)$ is known.
   - **t-Test:** Used when the population standard deviation $(\sigma)$ is unknown and estimated from the sample $(s)$.

2. **Sample Size:**
   - **Z-Test:** Typically used for large sample sizes (usually $(n \geq 30)$.
   - **t-Test:** Suitable for smaller sample sizes, especially when $(n < 30)$, but it becomes robust for larger samples.

3. **Distribution:**
   - **Z-Test:** Assumes a normal distribution of the population or a large sample size due to the Central Limit Theorem.
   - **t-Test:** More robust to deviations from normality, making it suitable for smaller samples.

4. **Critical Values:**
   - **Z-Test:** Critical values are obtained from the standard normal distribution.
   - **t-Test:** Critical values depend on the degrees of freedom and are obtained from the t-distribution.

**Example Scenarios:**
- **Z-Test:**
  - *Scenario:* You are conducting a study on the average height of adults in a country where the population standard deviation of height is well-known. You collect a large sample (e.g., $(n \geq 30)$ and want to test whether the average height of your sample differs significantly from the known population mean height.
  - *Test Type:* Z-Test for One Sample Mean.

- **t-Test:**
  - *Scenario:* You are conducting a study on the effectiveness of a new drug on a small group of patients (e.g., $(n < 30)$. You collect data on the blood pressure of patients before and after treatment and want to test whether there is a significant difference in mean blood pressure.
  - *Test Type:* Paired Sample t-Test.

In summary, use a Z-test when the population standard deviation is known, and you have a large sample size. Use a t-test when the population standard deviation is unknown, or the sample size is small, and you need to estimate the standard deviation from the sample.

### Q2: Differentiate between one-tailed and two-tailed tests.

**One-Tailed Test vs. Two-Tailed Test:**

1. **Definition:**
   - **One-Tailed Test:** A statistical test in which the critical region is on one side of the sampling distribution (either the right or left).
   - **Two-Tailed Test:** A statistical test in which the critical region is on both sides of the sampling distribution (both the right and left).

2. **Hypotheses:**
   - **One-Tailed Test:** Involves a directional hypothesis, specifying the expected direction of the effect (e.g., greater than, less than).
   - **Two-Tailed Test:** Involves a non-directional hypothesis, testing whether there is any significant difference without specifying the direction.

3. **Critical Region:**
   - **One-Tailed Test:** The critical region is located in one tail of the distribution.
   - **Two-Tailed Test:** The critical region is divided between both tails of the distribution.

4. **Alpha Level (Significance Level):**
   - **One-Tailed Test:** The significance level (\(\alpha\)) is typically divided by 2 (\(\alpha/2\)), allocating half to each tail.
   - **Two-Tailed Test:** The significance level (\(\alpha\)) is used as is, split equally between the two tails.

5. **Decision Rule:**
   - **One-Tailed Test:** Reject the null hypothesis if the test statistic falls into the critical region in the specified direction.
   - **Two-Tailed Test:** Reject the null hypothesis if the test statistic falls into either of the two critical regions.

6. **Example:**
   - **One-Tailed Test:** Testing whether a new drug increases the average performance (one tail for "greater than").
   - **Two-Tailed Test:** Testing whether a coin is biased (two tails for "different from").

7. **Use Cases:**
   - **One-Tailed Test:** When there is a specific directional hypothesis or when the focus is on detecting an effect in one direction.
   - **Two-Tailed Test:** When there is no specific expectation about the direction of the effect or when interest lies in detecting any significant difference.

In summary, the choice between a one-tailed and two-tailed test depends on the research question, the directional hypothesis, and the nature of the effect being investigated. One-tailed tests are more focused on a specific direction, while two-tailed tests are more inclusive, detecting differences in either direction.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

**Type I and Type II Errors in Hypothesis Testing:**

1. **Type I Error (False Positive):**
   - **Definition:** Occurs when the null hypothesis is incorrectly rejected when it is actually true.
   - **Symbol:** $( \alpha)$ (alpha) represents the probability of committing a Type I error.
   - **Consequence:** Falsely concluding that there is a significant effect or difference when there isn't one.

2. **Type II Error (False Negative):**
   - **Definition:** Occurs when the null hypothesis is not rejected when it is actually false.
   - **Symbol:** $( \beta)$ (beta) represents the probability of committing a Type II error.
   - **Consequence:** Failing to detect a significant effect or difference that truly exists.

**Example Scenarios:**

- **Type I Error (False Positive):**
  - **Scenario:** Suppose a medical researcher is testing a new drug to reduce cholesterol levels. The null hypothesis $(H_0)$ is that the drug has no effect on cholesterol, and the alternative hypothesis $(H_1)$ is that the drug reduces cholesterol. The researcher sets a significance level $(\alpha)$ of 0.05.
  - **Outcome:** If, based on the sample data, the researcher rejects $(H_0)$ and concludes that the drug is effective (when it's not), a Type I error has occurred. The researcher falsely believes the drug works.

- **Type II Error (False Negative):**
  - **Scenario:** Continuing with the example, suppose the drug actually does reduce cholesterol, but the sample size is small, and the effect is not detected. The researcher fails to reject $(H_0)$.
  - **Outcome:** A Type II error occurs. The researcher fails to recognize the drug's effectiveness, leading to the incorrect conclusion that the drug has no impact on cholesterol levels.

**Trade-Off between Type I and Type II Errors:**
- Adjusting the significance level $(\alpha)$ affects the likelihood of Type I and Type II errors.
- As $(\alpha)$ decreases (e.g., from 0.05 to 0.01), the risk of Type I error decreases, but the risk of Type II error may increase.

In summary, Type I and Type II errors represent the potential mistakes made in hypothesis testing. Researchers aim to balance these errors by choosing an appropriate significance level and considering the consequences of both types of errors in the context of the study.

### Q4: Explain Bayes's theorem with an example.

**Bayes's Theorem:**

Bayes's Theorem is a mathematical formula that describes the probability of an event, based on prior knowledge of conditions that might be related to the event. It is named after Thomas Bayes, an 18th-century statistician, and it plays a crucial role in Bayesian statistics.

The formula for Bayes's Theorem is given by:

$[ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)} ]$

where:
- $( P(A | B))$ is the probability of event A occurring given that event B has occurred (posterior probability).
- $( P(B | A))$ is the probability of event B occurring given that event A has occurred.
- $( P(A))$ and $( P(B))$ are the probabilities of events A and B, respectively.

**Example: Medical Diagnosis**

Let's consider a medical diagnosis scenario to illustrate Bayes's Theorem. Suppose we have a rare disease (event A) that affects 1 in 10,000 people $( P(A) = 0.0001)$. Additionally, there is a diagnostic test for this disease that is not perfect. The test correctly identifies 99% of people with the disease $( P(B | A) = 0.99 )$, but it also produces false positives for 2% of healthy individuals $( P(B | \neg A) = 0.02)$.

We want to find the probability that an individual has the disease given that the test result is positive $( P(A | B))$.

**Application of Bayes's Theorem:**

$[ P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B | A) \cdot P(A) + P(B | \neg A) \cdot P(\neg A)} ]$

Substitute the given values:

$[ P(A | B) = \frac{0.99 \cdot 0.0001}{0.99 \cdot 0.0001 + 0.02 \cdot 0.9999}]$

Calculate the result:

$[ P(A | B) \approx 0.0047 ]$

In this example, even if an individual tests positive for the disease, the probability of actually having the disease $( P(A | B)$ is relatively low due to the rarity of the disease and the imperfect nature of the test. This illustrates the importance of considering prior probabilities and the performance of the diagnostic test when interpreting test results.

### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

**Confidence Interval:**

A confidence interval is a statistical tool that provides a range of values within which we can reasonably expect the true parameter (such as a population mean or proportion) to fall. It is an interval estimate that quantifies the uncertainty associated with a sample statistic. The confidence interval is often expressed with a specified level of confidence, such as 95% or 99%.

**Calculating a Confidence Interval:**

The general formula for calculating a confidence interval for the mean $(\mu)$ of a population is given by:

$[ \text{Confidence Interval} = \bar{x} \pm \left( \text{Margin of Error} \right)]$

where:
- $(\bar{x})$ is the sample mean,
- $(\text{Margin of Error})$ is a critical value multiplied by the standard error of the mean.

The critical value is determined based on the chosen confidence level and the distribution (e.g., Z-score for a normal distribution, T-score for a t-distribution).

**Example: Confidence Interval for the Mean:**

Let's say we want to calculate a 95% confidence interval for the average height of a population. We collect a sample of 30 individuals and measure their heights. The sample mean (\(\bar{x}\)) is 65 inches, and the sample standard deviation (\(s\)) is 3 inches.

1. **Determine the Critical Value:**
   - If the sample size is large (typically \(n \geq 30\)), use the Z-score for a normal distribution.
   - For a 95% confidence interval, the Z-score is approximately 1.96.

2. **Calculate the Margin of Error:**
   $[ \text{Margin of Error} = Z \times \left( \frac{s}{\sqrt{n}} \right)]$
   $[ \text{Margin of Error} = 1.96 \times \left( \frac{3}{\sqrt{30}} \right)]$

3. **Calculate the Confidence Interval:**
   $[ \text{Confidence Interval} = \bar{x} \pm \text{Margin of Error}]$
   $[ \text{Confidence Interval} = 65 \pm 1.96 \times \left( \frac{3}{\sqrt{30}} \right)]$

   Calculate the lower and upper limits of the confidence interval.

**Interpretation:**
With a 95% confidence level, we can say that we are 95% confident that the true average height of the population falls within the calculated interval.

Keep in mind that the choice of distribution (normal or t-distribution) depends on the sample size and the assumptions made about the population. For smaller sample sizes, the t-distribution is often used.

### Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

**Bayes' Theorem Application: Probability Calculation**

Let's consider a scenario where we want to calculate the probability of a person having a certain medical condition given a positive test result. The relevant probabilities are as follows:

- The probability of having the medical condition: $( P(\text{Condition}) = 0.02)$
- The probability of testing positive given the condition: $( P(\text{Positive | Condition}) = 0.95 )$
- The probability of testing positive given not having the condition: $( P(\text{Positive | Not Condition}) = 0.10 )$

We want to find the probability of having the condition given a positive test result $( P(\text{Condition | Positive}) )$.

**Bayes' Theorem Formula:**
$[ P(\text{Condition | Positive}) = \frac{P(\text{Positive | Condition}) \cdot P(\text{Condition})}{P(\text{Positive | Condition}) \cdot P(\text{Condition}) + P(\text{Positive | Not Condition}) \cdot P(\text{Not Condition})}]$

**Solution:**
1. Calculate the probability of not having the condition:
   $[ P(\text{Not Condition}) = 1 - P(\text{Condition}) = 1 - 0.02 = 0.98 ]$

2. Substitute the given values into Bayes' Theorem:
   $[ P(\text{Condition | Positive}) = \frac{0.95 \cdot 0.02}{0.95 \cdot 0.02 + 0.10 \cdot 0.98}]$

3. Calculate the result:
   $[ P(\text{Condition | Positive}) \approx \frac{0.019}{0.019 + 0.098} \approx \frac{0.019}{0.117} \approx 0.162]$

**Interpretation:**
Given a positive test result, the probability that the person actually has the medical condition is approximately 0.162 or 16.2%. This highlights the importance of considering both the probability of the condition and the accuracy of the test when interpreting test results.

### Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.

**Calculating the 95% Confidence Interval:**

For a 95% confidence interval, we use the formula:

$[ \text{Confidence Interval} = \bar{x} \pm \left( \text{Margin of Error} \right)]$

where:
- $(\bar{x}) is the sample mean.
- $(\text{Margin of Error})$ is determined by the critical value multiplied by the standard error of the mean.

The formula for the margin of error is:

$[ \text{Margin of Error} = \text{Critical Value} \times \left( \frac{\text{Standard Deviation}}{\sqrt{\text{Sample Size}}} \right) ]$

For a 95% confidence interval with a normal distribution, the critical value is approximately 1.96.

**Given:**
- Sample Mean $(\bar{x})$ = 50
- Standard Deviation $(\sigma)$ = 5
- Sample Size $(n\)$ is not provided (assume a sufficiently large sample)

**Solution:**
1. **Determine the Margin of Error:**
   $[ \text{Margin of Error} = 1.96 \times \left( \frac{5}{\sqrt{n}} \right) ]$

2. **Substitute the Values and Solve for Sample Size (\(n\)):**
   $[ 1.96 \times \left( \frac{5}{\sqrt{n}} \right) = \text{Desired Margin of Error} ]$

   Solving this equation will give us the required sample size (\(n\)). As an example, let's assume $(n = 100)$.

3. **Calculate the Confidence Interval:**
   $[ \text{Confidence Interval} = 50 \pm \text{Margin of Error} ]$

**Interpretation:**
If the calculations yield $(n = 100)$ as an assumed sample size, then the 95% confidence interval for the population mean would be:

$[ \text{Confidence Interval} = 50 \pm \text{Calculated Margin of Error}]$

This means we are 95% confident that the true population mean falls within this interval. However, without the exact sample size, we cannot provide a specific numerical value for the confidence interval. The interpretation depends on the actual calculated margin of error based on the sample size.

### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

**Margin of Error in a Confidence Interval:**

The margin of error in a confidence interval represents the range within which we expect the true population parameter to lie. It is influenced by the variability in the data (standard deviation), the desired level of confidence, and the sample size. The formula for the margin of error is:

$[ \text{Margin of Error} = \text{Critical Value} \times \left( \frac{\text{Standard Deviation}}{\sqrt{\text{Sample Size}}} \right)]$

**Key Points:**
1. **Critical Value:** Determines how much the sample mean is likely to vary from the true population mean. It is based on the desired level of confidence (e.g., 95%, 99%) and the distribution (e.g., Z-score for a normal distribution, T-score for a t-distribution).
2. **Standard Deviation:** Represents the variability or spread of the data.
3. **Sample Size:** Appears in the denominator under the square root. Larger sample sizes lead to smaller standard errors and, consequently, smaller margins of error.

**Effect of Sample Size on Margin of Error:**
- As the sample size $(n)$ increases, the margin of error decreases.
- Larger sample sizes result in a more precise estimate of the population parameter.

**Example Scenario:**
Consider a scenario where researchers want to estimate the average height of a population with a 95% confidence level. They collect two different samples:
1. Sample A with 50 individuals.
2. Sample B with 500 individuals.

Assuming the same standard deviation for both samples, the larger sample size of Sample B would result in a smaller margin of error compared to Sample A. This increased precision is due to the larger sample size, allowing for a more accurate estimation of the population mean. In practical terms, a smaller margin of error indicates greater confidence in the accuracy of the sample mean as an estimate of the true population mean.

### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

**Calculating the Z-Score:**

The z-score is a measure of how many standard deviations a data point is from the mean of a distribution. The formula for calculating the z-score is:

$[ Z = \frac{X - \mu}{\sigma} ]$

where:
- $( X)$ is the individual data point.
- $( \mu )$ is the population mean.
- $( \sigma)$  is the population standard deviation.

**Given:**
- Data point $( X)$ = 75
- Population mean $( \mu )$ = 70
- Population standard deviation $( \sigma)$ = 5

**Solution:**
$[ Z = \frac{75 - 70}{5} = \frac{5}{5} = 1 ]$

**Interpretation:**
The z-score of 1 indicates that the data point (75) is 1 standard deviation above the mean (70) in the distribution. A positive z-score implies that the data point is above the mean, and the magnitude of the z-score (1 in this case) signifies how far it is from the mean in terms of standard deviations.

### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

**Hypothesis Testing for the Effectiveness of a Weight Loss Drug:**

Let's set up the hypothesis test to determine if the new weight loss drug is significantly effective. The hypotheses are as follows:

- **Null Hypothesis $(H_0)$:** The drug is not significantly effective. $( \mu = 0 )$
- **Alternative Hypothesis $(H_1)$:** The drug is significantly effective. $( \mu \neq 0)$

**Given:**
- Sample Mean $( \bar{x})$ = 6 pounds
- Sample Standard Deviation ( s )$ = 2.5 pounds
- Sample Size $( n )$ = 50
- Confidence Level = 95%

**Calculations:**
1. **Degrees of Freedom $( df )$:**
   $[ df = n - 1 = 50 - 1 = 49 ]$

2. **Critical Value:**
   For a two-tailed test at a 95% confidence level and 49 degrees of freedom, the critical t-value is obtained from a t-table or statistical software.

3. **Calculate the Standard Error $( SE)$:**
   $[ SE = \frac{s}{\sqrt{n}}]$

4. **Calculate the t-Statistic:**
   $[ t = \frac{\bar{x} - \mu_0}{SE}]$
   Here, $( \mu_0 )$ is the hypothesized population mean under the null hypothesis $( \mu_0 = 0)$.

5. **Compare the t-Statistic with Critical Value:**
   If the absolute value of the calculated t-statistic is greater than the critical t-value, we reject the null hypothesis.

**Interpretation:**
- If the calculated t-statistic falls in the critical region, we reject the null hypothesis, suggesting that the weight loss drug is significantly effective.
- If the calculated t-statistic does not fall in the critical region, we fail to reject the null hypothesis, indicating insufficient evidence to claim the drug's effectiveness.

Note: For precise calculations and interpretation, the critical t-value needs to be obtained from a t-table or statistical software. Additionally, ensure assumptions of normality and independence are met for valid results in hypothesis testing.

### Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

**Calculating the 95% Confidence Interval for Proportion:**

When estimating the confidence interval for a population proportion, we use the following formula:

$[ \text{Confidence Interval} = \hat{p} \pm Z \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} ]$

where:
- $(\hat{p})$ is the sample proportion.
- $(Z)$ is the critical z-value for the desired confidence level.
- $(n)$ is the sample size.

**Given:**
- Sample Proportion $(\hat{p})$ = 0.65 (65%)
- Sample Size $(n)$ = 500
- Confidence Level = 95%

**Calculations:**
1. **Critical Z-Value:**
   For a 95% confidence level, the critical z-value is approximately 1.96.

2. **Calculate the Margin of Error $(E)$:**
   $[E = Z \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}]$

3. **Calculate the Confidence Interval:**
   $[ \text{Confidence Interval} = \hat{p} \pm E ]$

**Interpretation:**
The 95% confidence interval for the true proportion of people satisfied with their job is calculated using the given sample data. This interval provides a range within which we are 95% confident that the true proportion in the entire population lies. The calculation involves considering the sample proportion, the critical z-value, and the sample size to determine the margin of error and construct the interval.

### Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

**Hypothesis Testing for Two Independent Samples:**

To determine if there is a significant difference in student performance between two teaching methods, we can conduct a two-sample t-test. The hypotheses are as follows:

- **Null Hypothesis $(H_0)$:** There is no significant difference in the mean scores of the two teaching methods. $( \mu_1 = \mu_2 )$
- **Alternative Hypothesis $(H_1)$:** There is a significant difference in the mean scores of the two teaching methods. $( \mu_1 \neq \mu_2 )$

**Given:**
- Sample A: Mean $( \bar{x}_1)$ = 85, Standard Deviation (\( s_1 \)) = 6, Sample Size (\( n_1 \)) is not provided.
- Sample B: Mean $( \bar{x}_2)$ = 82, Standard Deviation (\( s_2 \)) = 5, Sample Size (\( n_2 \)) is not provided.
- Significance Level $( \alpha)$ = 0.01

**Calculations:**
1. **Degrees of Freedom $( df)$:**
   $[ df = \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2 - 1}}]$

2. **Critical t-Value:**
   For a two-tailed test at a 99% confidence level and calculated degrees of freedom, obtain the critical t-value from a t-table or statistical software.

3. **Calculate the Standard Error $( SE )$:**
   $[ SE = \sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}} ]$

4. **Calculate the t-Statistic:**
   $[ t = \frac{\bar{x}_1 - \bar{x}_2}{SE}]$

5. **Compare the t-Statistic with Critical Value:**
   If the absolute value of the calculated t-statistic is greater than the critical t-value, reject the null hypothesis.

**Interpretation:**
- If the calculated t-statistic falls in the critical region, we reject the null hypothesis, indicating a significant difference in student performance between the two teaching methods.
- If the calculated t-statistic does not fall in the critical region, we fail to reject the null hypothesis, suggesting no significant difference in mean scores.

### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

To determine if caffeine has a significant effect on reaction time, we can conduct a one-sample t-test. The hypotheses are as follows:

- **Null Hypothesis $(H_0)$:** Caffeine does not have a significant effect on reaction time. $( \mu = \mu_0 )$
- **Alternative Hypothesis $(H_1)$:** Caffeine has a significant effect on reaction time. $( \mu \neq \mu_0 )$

- Sample Mean $( \bar{x})$ = 0.25 seconds
- Sample Standard Deviation $( s )$ = 0.05 seconds
- Sample Size $( n )$ = 30
- Confidence Level = 90%

**Calculations:**
1. **Degrees of Freedom $( df )$:**
   $[ df = n - 1 = 30 - 1 = 29]$

2. **Critical t-Value:**
   For a two-tailed test at a 90% confidence level and 29 degrees of freedom, obtain the critical t-value from a t-table or statistical software.

3. **Calculate the Standard Error $( SE )$:**
   $[ SE = \frac{s}{\sqrt{n}} ]$

4. **Calculate the t-Statistic:**
   $[ t = \frac{\bar{x} - \mu_0}{SE}]$
   Here, $( \mu_0)$ is the hypothesized population mean under the null hypothesis.

5. **Compare the t-Statistic with Critical Value:**
   If the absolute value of the calculated t-statistic is greater than the critical t-value, reject the null hypothesis.

**Interpretation:**
- If the calculated t-statistic falls in the critical region, we reject the null hypothesis, suggesting that caffeine has a significant effect on reaction time.
- If the calculated t-statistic does not fall in the critical region, we fail to reject the null hypothesis, indicating insufficient evidence to claim a significant effect of caffeine on reaction time.
