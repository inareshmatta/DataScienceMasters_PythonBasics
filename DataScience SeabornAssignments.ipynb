### 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
import pandas as pd
```

Seaborn is a powerful Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn along with their typical use cases:

1. **Bar Plot:**
   - **Use:** Display the distribution of a categorical variable or show the relationship between a categorical and a numerical variable.
   - **Example Use Case:** Visualizing the count of items in different categories.


2. **Histogram:**
   - **Use:** Display the distribution of a numerical variable.
   - **Example Use Case:** Visualizing the distribution of exam scores in a class.


3. **Scatter Plot:**
   - **Use:** Display the relationship between two numerical variables.
   - **Example Use Case:** Visualizing the correlation between the hours studied and exam scores.


4. **Box Plot:**
   - **Use:** Show the distribution of a numerical variable across different categories.
   - **Example Use Case:** Visualizing the distribution of salaries across different departments.

 
5. **Heatmap:**
   - **Use:** Visualize the correlation matrix of numerical variables in a dataset.
   - **Example Use Case:** Understanding the correlation between different features in a dataset.


These are just a few examples, and Seaborn offers many other plot types for various data visualization needs. Remember to customize the plots according to your specific project requirements.

### Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = pd.read_csv('fmri.csv')


# Plot a line plot
sns.set(style="darkgrid")  # Set the style of the plot

plt.figure(figsize=(10, 6))  # Set the size of the plot

# Create a line plot with different colors for events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Set plot labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time for Different Events and Regions")

# Display the legend
plt.legend(title="Event", loc="upper right")

# Show the plot
plt.show()
```

### 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = pd.read_csv('titanic.csv')

# Set the style of the plot
sns.set(style="whitegrid")

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Box plot for Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=titanic_data, ax=axes[0])
axes[0].set_title('Age Distribution by Pclass')

# Box plot for Fare vs Pclass
sns.boxplot(x='Pclass', y='Fare', data=titanic_data, ax=axes[1])
axes[1].set_title('Fare Distribution by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

### Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset from Seaborn
diamonds_data = pd.read_csv('diamonds.csv')

# Set the style of the plot
sns.set(style="whitegrid")

# Plot a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack", bins=30, kde=True)

# Set plot labels and title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")

# Display the legend
plt.legend(title="Cut", loc="upper right")

# Show the plot
plt.show()
```

### Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset from Seaborn
iris_data = pd.read_csv('iris.csv')

# Set the style of the plot
sns.set(style="ticks")

# Plot a pair plot with hue for the 'species' column
sns.pairplot(iris_data, hue='species')

# Show the plot
plt.show()
```

### Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset from Seaborn
flights_data = pd.read_csv("flights.csv")

# Pivot the dataset to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Set the style of the plot
sns.set(style="whitegrid")

# Plot a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=.5)

# Set plot labels and title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Number of Passengers (Flights Dataset)")

# Show the plot
plt.show()
```
